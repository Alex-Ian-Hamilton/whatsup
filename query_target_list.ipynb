```python
from astroquery.simbad import Simbad
```

```python
messier = ["M{0}".format(i) for i in range(1, 111)]
```

```python
Simbad.add_votable_fields('flux(V)')

mags = []

for i in range(len(messier)):
    mag = Simbad.query_object(messier[i])['FLUX_V']
    mags.append(mag[0])
```

```python
import numpy as np
from astropy.io import ascii
from astropy.table import Table

tbl = Table([messier, [m if type(m) is not np.ma.core.MaskedConstant else 'None' for m in mags]])

ascii.write(tbl, format='csv')
```
